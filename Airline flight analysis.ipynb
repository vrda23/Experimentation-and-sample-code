```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels
import matplotlib.pyplot as plt
import math
```

```python
plt.style.use('ggplot')

# Read in Data
flight = pd.read_csv(r"C:\Users\jvrdolja\Downloads\flight.csv")
display(flight.head())
```

```python
# Plotting a histogram for coach_price
price_hist= flight["coach_price"].plot(kind="hist",bins=30, title="Coach price histogram")
plt.show()
plt.clf()
```

```python
# Checking average price for coach seats
avg_coach = flight["coach_price"].mean()
print(avg_coach)
```

```python
## Subbseting df for flights that are 8 hours long
eight_hr = flight[flight["hours"]==8]
# Visualizing coach prices for 8-hour flights
eight_hist = eight_hr["coach_price"].plot(kind="hist",bins=30, title="8-hour Coach price histogram")
plt.show()
plt.clf()
```

```python
# Checking mean for 8-hour flights
print(eight_hr["coach_price"].mean())

# Checking how delay times are distributed
delays = sns.kdeplot(flight["delay"])
delays.set_title("Delay time distribution (min)")
plt.show()
plt.clf()
```

```python
# Checking the number of flights that were delayed more then an hour
print(flight[flight["delay"]>60].delay.count())

# Subbseting df for flights with longer delays
delays_over_hour = flight[flight["delay"]>60]
```

```python
# Visualizing delays longer than hour among days
long_delays= sns.countplot(delays_over_hour["day_of_week"])
long_delays.set_title("Num of delays per weekday")
long_delays.set_xlabel("Day of week")
long_delays.set_ylabel("Count")
plt.show()
plt.clf()
```

```python
# Visualizing the relationship between coach and first class prices
coach_first= sns.scatterplot(x=flight["coach_price"], y=flight["firstclass_price"])
coach_first.set_title("Coach vs First class relationship")
coach_first.set_xlabel("Coach price")
coach_first.set_ylabel("First class price")
plt.show()
plt.clf()
```

```python
# Visualizing flights with inflight Meals
h= sns.histplot(flight, x = "coach_price", hue = flight.inflight_meal)
h.set_xlabel("Coach price")
plt.show()
plt.clf()
```

```python
# Visualizing flights with flight entertainment
e =sns.histplot(flight, x = "coach_price", hue = flight.inflight_entertainment)
e.set_xlabel("Coach price")
plt.legend(title='Inflight entertainment', loc='upper left', labels=['Yes', 'No'])
plt.show()
plt.clf()
```
